```python
#series properties
#1) homogenous
#2) vectorized operations
#3) handle missing values with NaN
#4) mutable values , immutable size
```

```python
# vectorized operations( series properties)
s1= pd.Series([1,2,3,4,5])
s2= pd.Series([10,20,30,40,50])
print(s1+s2)
```

```python
# mutable values , immutable size means values of series can be change but size of series would still be same
s1= pd.Series([1,2,3,4,5])
s2= pd.Series([10,20,30,40,50])
s1[0]=100
changed_s1=s1.drop(0)  #drop() does NOT modify the original Series , ðŸ‘‰ It returns a new Series
print(s1)
print(changed_s1)
```
